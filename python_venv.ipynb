# Tutorial to create a pyhton enviroment 


A python environment is a simple way to work with multiple versions of Python on the same machine and same account. 
This is also useful to separate dependencies and allow users to configure their own Python environment without root or adminestrator access. 

## Purpose of this tutorial/howto: 
To allow you to 
1. set up a new environment

2. activate and deactivate the environment

3. install the environment (register it actually) with Jupyter (as a kernel)

4. where to find more information


## Where to find more information

1. official docs [https://docs.python.org/3/library/venv.html] 


## steps
first of all, create and change to the directory (with the `mkdir` and `cd` commands to your chosen work director)
In a terminal, which could also be inside the OMD platform (which is it self running inside a Jupyter environment)

here we create
```sh
$ mkdir learn_venv
$ cd learn_venv
$ python3 -m venv ./.venv # this creaes a folder .venv, which is hidden, so you need to use ls -a to see it (stand for all files) 
$ ls -lrtha
total 12
drwxrwxr-x 3 adham adham 4096 Jan 10 16:40 ./
drwxrwxr-x 3 adham adham 4096 Jan 10 16:39 ../
drwxrwxr-x 6 adham adham 4096 Jan 10 16:40 .venv/

```
- as we see this has created the folder `.venv` which contains:

```sh
.
└── venv_learn_ase
    ├── bin
    ├── include
    ├── lib
    ├── lib64 -> lib
    ├── pyvenv.cfg
    └── share
```

now we need to **activate** the environment with (do not forget the 'dot' which is part of the name: 

```bash
$ source .venv/bin/activate 
 
```

ands this activates your new, prestine python environment `(.venv)`. It adds to your terminal prompt the `(.venv)`, which woud look like so for example: 
```
(.venv) user@ubuntu:~/learn_venv$  
```


*Note: The built-in bash source command reads and executes the content of a file.*

## Installing/registering the env with Jupyter

run the following while in the activated environment: 
```
$ python3 -m ipykernel install --user --name=learn_venv_dot_venv
```
note that you can give the kernel any name you want, usually it should be something simple.
